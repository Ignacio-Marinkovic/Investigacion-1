```python
import numpy as np
from matplotlib import pyplot as plt

from multimeter.utils.utils import (
    n_sigma_to_PTE,
    PTE_to_L,
    L_iso_to_n_eff,
)

import git
import os

# Path to the root of the git repository
path_git = git.Repo('.', search_parent_directories=True).working_tree_dir
path_global = os.path.dirname(path_git)
print(f'Path to git repo: {path_global}')


def N_sigma_to_L_and_N_eff(N_sigma,Nd):
    PTE = n_sigma_to_PTE(N_sigma)
    L = PTE_to_L(PTE=PTE, Nd=Nd)
    N_eff = L_iso_to_n_eff(d=2, L=L)
    return L, N_eff

```

## DESI DR1 vs DESI DR2

```python
#DESI DR1 N_Sigma
N_sigma = 1.9
L, N_eff = N_sigma_to_L_and_N_eff(N_sigma, Nd=2)

print(f'DESI DR1 N_sigma: {N_sigma:.3g}')
print(f'DESI DR1 L: {L:.3g}')
print(f'DESI DR1 N_eff: {N_eff:.3g}')
print('---')

#DESI DR2 N_Sigma
N_sigma = 2.3
L, N_eff = N_sigma_to_L_and_N_eff(N_sigma, Nd=2)

print(f'DESI DR2 N_sigma: {N_sigma:.3g}')
print(f'DESI DR2 L: {L:.3g}')
print(f'DESI DR2 N_eff: {N_eff:.3g}')
```

```python
N_sigma_values = np.linspace(0, 7, 200)
N_eff_values_2n = np.zeros(len(N_sigma_values))
L_2n = np.zeros(len(N_sigma_values))
L_3n = np.zeros(len(N_sigma_values))
for i, N_sigma in enumerate(N_sigma_values):
    L, N_eff = N_sigma_to_L_and_N_eff(N_sigma, Nd=2)
    L_2n[i] = L
    N_eff_values_2n[i] = N_eff

N_eff_values_3n = np.zeros(len(N_sigma_values))
for i, N_sigma in enumerate(N_sigma_values):
    L, N_eff = N_sigma_to_L_and_N_eff(N_sigma, Nd=3)
    N_eff_values_3n[i] = N_eff
    L_3n[i] = L
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

# -------------------------
# PRL-like rcParams + "LaTeX-like" fonts (sin usetex)
# -------------------------
mpl.rcParams.update({
    "figure.dpi": 300,
    "savefig.dpi": 300,

    # Fuente tipo LaTeX sin depender de TeX instalado
    "font.family": "serif",
    "mathtext.fontset": "cm",      # Computer Modern
    "mathtext.rm": "serif",

    # Si tenés LaTeX instalado y querés render real:
    # "text.usetex": True,

    "font.size": 10,
    "axes.labelsize": 11,
    "axes.titlesize": 11,
    "legend.fontsize": 9,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "axes.linewidth": 0.8,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "xtick.minor.size": 2,
    "ytick.minor.size": 2,
    "xtick.major.width": 0.8,
    "ytick.major.width": 0.8,
    "xtick.minor.width": 0.6,
    "ytick.minor.width": 0.6,
    "legend.frameon": False,
})

# -------------------------
# Helpers
# -------------------------
x_h_black = y1
x_h_blue  = np.interp(y2, N_eff_values_2n, N_sigma_values)

# -------------------------
# Figure + palette (Okabe–Ito)
# -------------------------
fig, ax = plt.subplots(figsize=(3.37, 2.6), constrained_layout=True)

CB_BLACK  = '#000000'
CB_BLUE   = '#0072B2'
CB_ORANGE = '#E69F00'
CB_PURPLE = '#CC79A7'

# Curvas principales
ax.plot(N_sigma_values, N_sigma_values, '--', color=CB_BLACK, lw=1.2,
        label=r'$N=2,\ D=1$')

ax.plot(N_sigma_values, N_eff_values_2n, color=CB_BLUE, lw=1.4,
        label=r'$N=2,\ D=2$')

#ax.plot(N_sigma_values, N_eff_values_3n, color=CB_BLUE, lw=1.4,
#        label=r'$N=2,\ D=2$')


# Guías
ax.vlines(x1, 0, y1, linestyles=':', colors=CB_BLACK,  lw=1.0, alpha=0.9)
ax.vlines(x2, 0, y2, linestyles=':', colors=CB_ORANGE, lw=1.0, alpha=0.9)
ax.hlines(y2, 0, x_h_blue, linestyles=':', colors=CB_PURPLE, lw=1.0, alpha=0.9)

# Punto en la intersección (naranja con magenta): (x2, y2)
ax.plot(x2, y2, marker='o', ms=5.5, mfc=CB_PURPLE, mec=CB_ORANGE, mew=1.1, zorder=5)

# -------------------------
# Textos con valores (en el color correspondiente)
# -------------------------
# Vertical negra (x1)
ax.text(x1 - 0.3, 0.35, rf'$N_\sigma(H_0)={x1:.2f}$', color=CB_BLACK, fontsize=7,
        rotation=90, va='bottom', ha='left')

# Vertical naranja (x2) y su y2
ax.text(x2 + 0.12, 0.35, rf'$N_\sigma(\Omega_m\!-\!H_0)={x2:.2f}$', color=CB_ORANGE, fontsize=7,
        rotation=90, va='bottom', ha='left')

# Horizontal magenta (y2) y el x donde corta la curva azul
ax.text(0.20, y2 - 0.5,
        rf'$N_\sigma^{{\rm eff}}(\Omega_m\!-\!H_0)={y2:.2f}$',
        color=CB_PURPLE, fontsize=7, va='bottom', ha='left')

# (Opcional) etiqueta cerca del punto de intersección
#ax.text(x2 + 0.18, y2 + 0.18, rf'$({x2:.2f},{y2:.2f})$', color=CB_PURPLE, fontsize=7)

# -------------------------
# Leyenda: handles extra
# -------------------------
vline_handle_H0 = Line2D([], [], color=CB_BLACK, linestyle=':', lw=1.0, label=r'$N_\sigma(H_0)$')
vline_handle_OmH0 = Line2D([], [], color=CB_ORANGE, linestyle=':', lw=1.0, label=r'$N_\sigma(\Omega_m\!-\!H_0)$')
vline_handle_horizontal = Line2D([], [], color=CB_PURPLE, linestyle=':', lw=1.0,
                                 label=r'$N_\sigma^{\rm eff}(\Omega_m\!-\!H_0)$')

handles, labels = ax.get_legend_handles_labels()
handles.extend([vline_handle_H0, vline_handle_OmH0, vline_handle_horizontal])
ax.legend(handles=handles, loc='upper left', fontsize=6)

# Labels
ax.set_xlabel(r'$N_\sigma$')
ax.set_ylabel(r'$N_\sigma^{\rm eff}$')
#ax.set_title(r'Planck18 / PPS tension')

# Ticks enteros (sin saltear)
ax.set_xlim(0, 7)
ax.set_ylim(0, 7)
ax.set_xticks(np.arange(0, 8, 1))
ax.set_yticks(np.arange(0, 8, 1))
ax.minorticks_on()

# Grid sutil
ax.grid(True, which='major', alpha=0.15, linewidth=0.6)
ax.grid(False, which='minor')

# Spines
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)

# Guardado
os.chdir(os.path.join(path_global, 'multimeter', 'notebooks', 'figures'))
fig.savefig('N_sigma_vs_N_eff.pdf', bbox_inches='tight')
fig.savefig('N_sigma_vs_N_eff.png', bbox_inches='tight')
plt.show()
```
